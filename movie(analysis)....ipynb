```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sms
```

```python
data = pd.read_csv('movie_sales.csv.csv')
```

```python
print(data)
```

```python
data
```

```python
data.head() ## checking for the hearders
```

```python
## checking for number of columns and column names

data.columns
```

```python
## to check for missing values

data.isnull().sum
```

```python
## droping the description column

data = data.drop('Description', axis=1)
```

```python
data
```

```python
data = data.drop('Director', axis=1)
```

```python
data = data.drop('Rank', axis=1) 
```

```python
data = data.drop('Actors', axis=1) 
```

```python
data
```

```python
## replacing missing value in revenue column

mean_value = data['Revenue (Millions)'].mean()
mean_value
```

```python
## replacing the missing value

data['Revenue (Millions)'].fillna(mean_value, inplace=True)
```

```python
## to confirm the replace

data.isnull().sum
```

```python
## replacing missing value in metascore column

mean_value = data['Metascore'].mean()
mean_value
```

```python
## replacing the missing value in metascore

data['Metascore'].fillna(mean_value, inplace=True)
```

```python
data.isnull().sum
```

```python
## To check for duplicate 
data.duplicated()
```

```python
## What is the total revenue generated by Genres?

revenue_by_genre = data.groupby('Genre')['Revenue (Millions)'].sum()
revenue_by_genre.round(0)[:10]
```

```python
plt.pie(revenue_by_genre.round(0)[:10])
plt.show
```

```python
## What is the total number of movies in the dataset

total_movie = data['Title'].count()
total_movie.astype(int)
```

```python
## What is the number of movies produced by year?

movies_per_year = data.groupby('Year')['Title'].count()
movies_per_year
```

```python
plt.pie(movies_per_year)
plt.show
```

```python
## What is the total revenue generated?

total_revenue = data['Revenue (Millions)'].sum()
total_revenue.astype(int)
```

```python
## Top 5 movies by rating

selected_col = data[['Title', 'Rating']][0:6]
selected_col
```

```python
plt.plot(selected_col)
plt.show
```

```python
##  Find the top 5 profitable movies.

profitable_movies = data[['Title', 'Revenue (Millions)']][0:6]
profitable_movies
```

```python
## Find 5 most voted movies

voted = data[['Title', 'Votes']][0:5]
voted
```

```python
##  Show the number of Genres by percentage of total

genre_percentage = revenue_by_genre / revenue_by_genre.sum() *100
genre_percentage.round(2)[0:10]
```

```python
## saving the data as a csv file

data.to_csv('movie_analysis.csv')
```
